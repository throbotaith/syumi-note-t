#### Affine setsと Convex sets

$y = \theta * x_1 + (1 - \theta)*x_2$

を考えてみます。下のコードを実行してみると,thetaの値が変化することでそれに対応した点たちが直線となることがわかります。この直線内の各点の集合を**アフィン集合** といいます。

一方で,$ 0 \leq \theta \leq 1$　の制約を持って作られた線分上の各点の集合を**凸集合(Convec set)** といいます。

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#以下のようなthetaをもつ関数を考えます.

x1 = -0.2
x2 = 0.2

def form(theta,x1,x2):
    return theta * x1 + (1 - theta) * x2

thetas = []
for i in range(-200,200,2):
     thetas.append(i * 0.01)
y = []
for theta in thetas:
    y.append(form(theta,x1,x2))

plt.plot(thetas, y, label="test")



# 凡例の表示
plt.legend("a")


# プロット表示(設定の反映)
plt.show()
thetas_convex = []
for i in range(-100,100,2):
     thetas_convex.append(i * 0.01)
y_convex = []
for theta in thetas_convex:
    y_convex.append(form(theta,x1,x2))

plt.plot(thetas_convex, y_convex, label="test")
plt.plot(thetas_convex, y_convex, label="test")



# 凡例の表示
plt.legend("c")

# プロット表示(設定の反映)
plt.show()


```
