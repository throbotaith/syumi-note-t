```python
a  = np.ndarray([1,0,3])
b = np.ndarray([2,0,1])


# check hyperplanes

def hyperplanes(a, b):
    def check_hyperplane(x):
        return np.dot(a, x) == b

    return check_hyperplane

a = np.array([1,2])
b = 5
hp = hyperplanes(a,b)
x1 = np.array([1, 2])
x2 = np.array([2, 3])

print(hp(x1))
print(hp(x2))
```

```python
#超平面の描写
import numpy as np
import matplotlib.pyplot as plt

def plot_hyperplane(a, b):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # データ点の作成
    x = np.linspace(-10, 10, 100)
    y = np.linspace(-10, 10, 100)
    X, Y = np.meshgrid(x, y)

    # 超平面の計算
    Z = (b - a[0] * X - a[1] * Y) / a[2]

    # 超平面の描画
    ax.plot_surface(X, Y, Z, alpha=0.5)

    # 軸ラベルの設定
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    # グラフの表示
    plt.show()

# 超平面の法線ベクトルと定数項の設定
a = np.array([1, -2, 1])
b = 5

# 超平面の描画
plot_hyperplane(a, b)
```

```python
#Euclidean balls
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot_euclidean_ball(xc, r):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # データ点の作成
    theta = np.linspace(0, 2 * np.pi, 100)
    phi = np.linspace(0, np.pi, 100)
    Theta, Phi = np.meshgrid(theta, phi)
    x = xc[0] + r * np.sin(Phi) * np.cos(Theta)
    y = xc[1] + r * np.sin(Phi) * np.sin(Theta)
    z = xc[2] + r * np.cos(Phi)

    # ユークリッドボールの描画
    ax.plot_surface(x, y, z, alpha=0.5)

    # 軸ラベルの設定
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    # グラフの表示
    plt.show()

def plot_ellipsoid(xc, A):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # データ点の作成
    u = np.linspace(0, 2 * np.pi, 100)
    v = np.linspace(0, np.pi, 100)
    U, V = np.meshgrid(u, v)
    x = xc[0] + np.sqrt(1 / A[0, 0]) * np.cos(U) * np.sin(V)
    y = xc[1] + np.sqrt(1 / A[1, 1]) * np.sin(U) * np.sin(V)
    z = xc[2] + np.sqrt(1 / A[2, 2]) * np.cos(V)

    # 楕円体の描画
    ax.plot_surface(x, y, z, alpha=0.5)

    # 軸ラベルの設定
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    # グラフの表示
    plt.show()

# ユークリッドボールの中心座標と半径
xc = np.array([1, 2, 3])
r = 5

# ユークリッドボールの描画
plot_euclidean_ball(xc, r)

# 楕円体の中心座標と共分散行列
xc = np.array([1, 2, 3])
A = np.array([[4, 0, 0], [0, 1, 0], [0, 0, 1]])

# 楕円体の描画
plot_ellipsoid(xc, A)
```

```python
#凸円錐
import numpy as np
import matplotlib.pyplot as plt

def plot_convex_cone(vertices):
    fig, ax = plt.subplots()

    # 凸錐を形成する頂点を結ぶ線分を描画
    num_vertices = len(vertices)
    for i in range(num_vertices):
        v1 = vertices[i]
        v2 = vertices[(i + 1) % num_vertices]
        ax.plot([v1[0], v2[0]], [v1[1], v2[1]], 'k-')

    # 頂点をプロット
    xs = vertices[:, 0]
    ys = vertices[:, 1]
    ax.plot(xs, ys, 'ro')

    # グラフの表示
    plt.show()

# 凸錐を形成する頂点の座標
vertices = np.array([[0, 0],
                     [1, 0],
                     [0.5, 1]])

# 凸錐の描画
plot_convex_cone(vertices)
```
